```
img = cv2.resize(image,(200,200))#for resizing the picture
cv2_imshow(img)
```

```
rows, cols = image.shape[:2]
M = cv2.getRotationMatrix2D((cols/2,rows/2),180,1)#jsut change the angle for rotation
```

```
dst = cv2.warpAffine(image,M,(cols,rows))
```

```
cv2_imshow(dst)
```
